```python
 # Pandas Series by using the command pd.Series(data, index), where index is a list of index labels. Let's use a Pandas Series to store a grocery list.
 
 groceries = pd.Series(data = [30, 6, 'Yes', 'No'], index = ['eggs', 'apples', 'milk', 'bread'])
 
```

```python
# We check whether bananas is a food item (an index) in Groceries
x = 'bananas' in groceries
```

he attribute .loc stands for location and it is used to explicitly state that we are using a labeled index. Similarly, the attribute .iloc stands for integer location and it is used to explicitly state that we are using a numerical index

```python
print('How many eggs and apples do we need to buy:\n',  groceries[[0, 1]]) 

# How many eggs and apples do we need to buy:
# eggs       30
# apples     6
# dtype: object

print('Do we need bread:\n', groceries[[-1]]) 
# Do we need bread:
# bread     No
# dtype: object

print('How many eggs do we need to buy:', groceries[0]) 
# How many eggs do we need to buy: 30

print('Do we need milk and bread:\n', groceries.iloc[[2, 3]]) 
# Do we need milk and bread:
# milk       Yes
# bread     No
# dtype: object
```

```python
import pandas as pd

# Pandas DataFrames are two-dimensional data structures with labeled rows and columns, that can hold many data types. 
# If you are familiar with Excel, you can think of Pandas DataFrames as being similar to a spreadsheet. 

items = {'Bob' : pd.Series(data = [245, 25, 55], index = ['bike', 'pants', 'watch']),
         'Alice' : pd.Series(data = [40, 110, 500, 45], index = ['book', 'glasses', 'bike', 'pants'])}

# print(type(items))
# class 'dict'

# We create a Pandas DataFrame by passing it a dictionary of Pandas Series
shopping_carts = pd.DataFrame(items)

shopping_carts

# tabular form,
# row labels == union of the index labels, 
# column labels == keys of the dictionary
# columns are arranged alphabetically

# We create a dictionary of lists (arrays)
data = {'Integers' : [1,2,3],
        'Floats' : [4.5, 8.2, 9.6]}

# We create a DataFrame 
df = pd.DataFrame(data)

# df
shopping_carts
```

```python
# We create a list of Python dictionaries
items2 = [{'bikes': 20, 'pants': 30, 'watches': 35}, 
          {'watches': 10, 'glasses': 50, 'bikes': 15, 'pants':5}]

# We create a DataFrame 
store_items = pd.DataFrame(items2)

# We display the DataFrame
store_items

# Again, notice that since the items2 dictionary we created doesn't have label indices, Pandas automatically uses numerical row indexes when it creates the DataFrame
```

```python
# accessing individual elements in a DataFrame
# dataframe[column][row]

store_items['glasses'][1] = 20.2

# We add a new column named shirts to our store_items DataFrame indicating the number of
# shirts in stock at each store. We will put 15 shirts in store 1 and 2 shirts in store 2
store_items['shirts'] = [15,2]


store_items
```

```python
store_items['suits'] = store_items['pants'] + store_items['shirts']

store_items
```
