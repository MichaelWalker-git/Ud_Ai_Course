```python
 # Pandas Series by using the command pd.Series(data, index), where index is a list of index labels. Let's use a Pandas Series to store a grocery list.
 
 groceries = pd.Series(data = [30, 6, 'Yes', 'No'], index = ['eggs', 'apples', 'milk', 'bread'])
 
```

```python
# We check whether bananas is a food item (an index) in Groceries
x = 'bananas' in groceries
```

he attribute .loc stands for location and it is used to explicitly state that we are using a labeled index. Similarly, the attribute .iloc stands for integer location and it is used to explicitly state that we are using a numerical index

```python
print('How many eggs and apples do we need to buy:\n',  groceries[[0, 1]]) 

# How many eggs and apples do we need to buy:
# eggs       30
# apples     6
# dtype: object

print('Do we need bread:\n', groceries[[-1]]) 
# Do we need bread:
# bread     No
# dtype: object

print('How many eggs do we need to buy:', groceries[0]) 
# How many eggs do we need to buy: 30

print('Do we need milk and bread:\n', groceries.iloc[[2, 3]]) 
# Do we need milk and bread:
# milk       Yes
# bread     No
# dtype: object
```

```python
# Panda series are mutable like NumPy ndarrays

# We change the number of eggs to 2
groceries['eggs'] = 2

# We can also delete items from a Pandas Series by using the .drop() method. The Series.drop(label) 
print('We remove apples (out of place):\n', groceries.drop('apples'))

# We remove apples from our grocery list in place by setting the inplace keyword to True
groceries.drop('apples', inplace = True)
# Arithmetic Operations on Pandas Series
fruits= pd.Series(data = [10, 6, 3,], index = ['apples', 'oranges', 'bananas'])

print('fruits + 2:\n', fruits + 2) # We add 2 to each item in fruits

print('fruits - 2:\n', fruits - 2) # We subtract 2 to each item in fruits

print('fruits * 2:\n', fruits * 2) # We multiply each item in fruits by 2 

print('fruits / 2:\n', fruits / 2) # We divide each item in fruits by 2

print('EXP(X) = \n', np.exp(fruits))

print('SQRT(X) =\n', np.sqrt(fruits))

print('POW(X,2) =\n',np.power(fruits,2)) # We raise all elements of fruits to the power of 2

# We print fruits for reference
print('Original grocery list of fruits:\n ', fruits)


# We add 2 only to the bananas
print('Amount of bananas + 2 = ', fruits['bananas'] + 2)


# We subtract 2 from apples
print('Amount of apples - 2 = ', fruits.iloc[0] - 2)


# We multiply apples and oranges by 2
print('We double the amount of apples and oranges:\n', fruits[['apples', 'oranges']] * 2)

# We divide apples and oranges by 2
print('We half the amount of apples and oranges:\n', fruits.loc[['apples', 'oranges']] / 2)
```

```python
import pandas as pd

# Pandas DataFrames are two-dimensional data structures with labeled rows and columns, that can hold many data types. 
# If you are familiar with Excel, you can think of Pandas DataFrames as being similar to a spreadsheet. 

items = {'Bob' : pd.Series(data = [245, 25, 55], index = ['bike', 'pants', 'watch']),
         'Alice' : pd.Series(data = [40, 110, 500, 45], index = ['book', 'glasses', 'bike', 'pants'])}

# print(type(items))
# class 'dict'

# We create a Pandas DataFrame by passing it a dictionary of Pandas Series
shopping_carts = pd.DataFrame(items)

shopping_carts

# tabular form,
# row labels == union of the index labels, 
# column labels == keys of the dictionary
# columns are arranged alphabetically

# We create a dictionary of lists (arrays)
data = {'Integers' : [1,2,3],
        'Floats' : [4.5, 8.2, 9.6]}

# We create a DataFrame 
df = pd.DataFrame(data)

# df
shopping_carts
```

```python
# We create a list of Python dictionaries
items2 = [{'bikes': 20, 'pants': 30, 'watches': 35}, 
          {'watches': 10, 'glasses': 50, 'bikes': 15, 'pants':5}]

# We create a DataFrame 
store_items = pd.DataFrame(items2)

# We display the DataFrame
store_items

# Again, notice that since the items2 dictionary we created doesn't have label indices, Pandas automatically uses numerical row indexes when it creates the DataFrame
```

```python
# accessing individual elements in a DataFrame
# dataframe[column][row]

store_items['glasses'][1] = 20.2

# We add a new column named shirts to our store_items DataFrame indicating the number of
# shirts in stock at each store. We will put 15 shirts in store 1 and 2 shirts in store 2
store_items['shirts'] = [15,2]


store_items
```

```python
store_items['suits'] = store_items['pants'] + store_items['shirts']

store_items
```

```python
#  add rows to our DataFrame
new_items = [{'bikes': 20, 'pants': 30, 'watches': 35, 'glasses': 4}]

new_store = pd.DataFrame(new_items, index = ['store 3'])

new_store

store_items = store_items.append(new_store, sort=True)

store_items
```

```python
# We add a new column using data from particular rows in the watches column
store_items['new watches'] = store_items['watches'][1:]

store_items

```

```python

# We insert a new column with label shoes right before the column with numerical index 4
store_items.insert(4, 'shoes', [8,5,0])
```

To delete rows and columns from our DataFrame we will use the .pop() and .drop() methods. The .pop() method only allows us to delete columns, while the .drop() method can be used to delete both rows and columns by use of the axis keyword

```python
store_items.pop('new watches')

# We remove the watches and shoes columns

store_items = store_items.drop(['watches', 'shoes'], axis = 1)

# We remove the store 2 and store 1 rows
store_items = store_items.drop(['store 2', 'store 1'], axis = 0)

# We change the column label bikes to hats
store_items = store_items.rename(columns = {'bikes': 'hats'})

# We change the row label from store 3 to last store
store_items = store_items.rename(index = {'store 3': 'last store'})

# We change the row index to be the data in the pants column
store_items = store_items.set_index('pants')
```

```python
## Dealing with NaN
# This means we need to have a method for detecting and correcting errors in our data.

items2 = [{'bikes': 20, 'pants': 30, 'watches': 35, 'shirts': 15, 'shoes':8, 'suits':45},
{'watches': 10, 'glasses': 50, 'bikes': 15, 'pants':5, 'shirts': 2, 'shoes':5, 'suits':7},
{'bikes': 20, 'pants': 30, 'watches': 35, 'glasses': 4, 'shoes':10}]


# We create a DataFrame  and provide the row index
store_items = pd.DataFrame(items2, index = ['store 1', 'store 2', 'store 3'])

store_items

```

```python
# We print the number of non-NaN values in our DataFrame
print()
print('Number of non-NaN values in the columns of our DataFrame:\n', store_items.count())
```

```python
# What to do with NaN values?
# Will elimante any rows with NaN values
store_items.dropna(axis = 0)
```

```python
store_items.dropna(axis = 1)
# Only shows complete columns
```

```python
# We replace all NaN values with 0
store_items.fillna(0)
```

```python
# We replace NaN values with the previous value in the column
store_items.fillna(method = 'ffill', axis = 0)
```

```python
# We replace NaN values with the previous value in the row
store_items.fillna(method = 'ffill', axis = 1)
```

```python
# We replace NaN values with the next value in the column
store_items.fillna(method = 'backfill', axis = 0)

```

```python
# All above methods are outPlace 
# inplace = True inside of fillna()

# .interpolate(method = 'linear', axis)
# method will use linear interpolation to replace NaN values using the values along the given axis.
store_items.interpolate(method = 'linear', axis = 0)
```

```python
import pandas as pd
import numpy as np

# Since we will be working with ratings, we will set the precision of our 
# dataframes to one decimal place.
pd.set_option('precision', 1)

# Create a Pandas DataFrame that contains the ratings some users have given to a
# series of books. The ratings given are in the range from 1 to 5, with 5 being
# the best score. The names of the books, the authors, and the ratings of each user
# are given below:

books = pd.Series(data = ['Great Expectations', 'Of Mice and Men', 'Romeo and Juliet', 'The Time Machine', 'Alice in Wonderland' ])
authors = pd.Series(data = ['Charles Dickens', 'John Steinbeck', 'William Shakespeare', ' H. G. Wells', 'Lewis Carroll' ])

user_1 = pd.Series(data = [3.2, np.nan ,2.5])
user_2 = pd.Series(data = [5., 1.3, 4.0, 3.8])
user_3 = pd.Series(data = [2.0, 2.3, np.nan, 4])
user_4 = pd.Series(data = [4, 3.5, 4, 5, 4.2])

# Users that have np.nan values means that the user has not yet rated that book.
# Use the data above to create a Pandas DataFrame that has the following column
# labels: 'Author', 'Book Title', 'User 1', 'User 2', 'User 3', 'User 4'. Let Pandas
# automatically assign numerical row indices to the DataFrame. 

# Create a dictionary with the data given above
dat = {
    'Author': authors, 
    'Book Title': books, 
    'User 1': user_1, 
    'User 2': user_2, 
    'User 3': user_3, 
    'User 4': user_4
}

data = [authors, books, user_1, user_2, user_3, user_4]

# Use the dictionary to create a Pandas DataFrame
book_ratings = pd.DataFrame(dat)

# If you created the dictionary correctly you should have a Pandas DataFrame
# that has column labels: 'Author', 'Book Title', 'User 1', 'User 2', 'User 3',
# 'User 4' and row indices 0 through 4.

# Now replace all the NaN values in your DataFrame with the average rating in
# each column. Replace the NaN values in place. HINT: you can use the fillna()
# function with the keyword inplace = True, to do this. Write your code below:
book_ratings.fillna(book_ratings.mean(), inplace=True)

```
